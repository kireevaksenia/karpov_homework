5. Посчитайте число заказов для каждого пользователя (CustomerID) из Германии (Germany). 
Оставьте только тех, кто совершил более N транзакций (InvoiceNo), где N – 80% процентиль. 
Запишите полученные id пользователей в germany_top (не весь датафрейм, только id).

```python
germany = retail.query('Country == "Germany"').groupby('CustomerID').agg({'InvoiceNo': 'nunique'})
q = germany.quantile(q=0.8)[0]
germany = germany.query('InvoiceNo > @q').reset_index()
germany_top = germany.CustomerID
germany
```

6. Используя объект с id пользователей (germany_top), полученный на предыдущем шаге, 
отфильтруйте наблюдения и оставьте в данных записи только по интересующим нас юзерам. 
Результирующий датафрейм запишите в top_retail_germany.

```python
top_retail_germany = retail[retail.CustomerID.isin(germany_top)]
```

7. Сгруппируйте top_retail_germany по коду товара (StockCode). Какой из продуктов добавляли в корзину чаще всего, кроме POST? 

Note: одним заказом считается единовременная покупка любого количества товара, т.е. без учета Quantity

```python
top_retail_germany.groupby('StockCode').agg({'InvoiceNo': 'nunique'}).sort_values('InvoiceNo')
```

8. Вернемся к анализу полного датасета retail. Создайте колонку Revenue с суммой покупки, используя колонки Quantity и UnitPrice.

```python
retail['Revenue'] = retail.Quantity * retail.UnitPrice
retail
```

9. Для каждой транзакции (InvoiceNo), посчитайте финальную сумму заказа. В качестве ответа укажите топ-5 (InvoiceNo) по сумме заказа (через запятую с пробелом и в порядке убывания TotalRevenue)
