### 3. Бутстрап средних

```python
boot = 100
avg = []
conf_level = 0.95
```

```python
for i in range(boot):
    samples = test.sample(len(test), replace=True)
    avg.append(samples.mean())

test.mean(), np.mean(avg)
```

```python
left_board = (1 - conf_level)/2
right_board = 1 - (1 - conf_level)/2
```

```python
test.quantile([left_board, right_board])
```

```python
def get_bootstrap(data_1, data_2, boot_it=1000, statistic=np.mean, conf_level=0.95):
    
    boot_data = []
    boot_len = max([len(data_1), len(data_2)])
    
    for i in tqdm(range(boot_it)):
        samples_1 = data_1.sample(len(data_1), replace=True).values
        samples_2 = data_2.sample(len(data_2), replace=True).values

        boot_data.append(statistic(samples_1 - samples_2))
        
    pd_boot_data = pd.DataFrame(boot_data)
    
    left_board = (1 - conf_level)/2
    right_board = 1 - (1 - conf_level)/2
    
    ci = pd_boot_data.quantile([left_board, right_board])
    
    p_1 = ss.norm.cdf(
            x=0,
            loc=np.mean(boot_data),
            scale=np.std(boot_data))
    
    p_2 = ss.norm.cdf(
            x=0,
            loc=-np.mean(boot_data),
            scale=np.std(boot_data))
    
    p_value = min(p_1, p_2) * 2
    
    _, _, bars = plt.hist(pd_boot_data[0], bins = 50)
    for bar in bars:
        if bar.get_x() <= ci.iloc[0][0] or bar.get_x() >= ci.iloc[1][0]:
            bar.set_facecolor('red')
        else: 
            bar.set_facecolor('grey')
            bar.set_edgecolor('black')
    
    plt.style.use('ggplot')
    plt.vlines(ci,ymin=0,ymax=50,linestyle='--')
    plt.xlabel('boot_data')
    plt.ylabel('frequency')
    plt.title("Histogram of boot_data")
    plt.show()
    
    
    return {'boot_data': boot_data,
           'ci': ci,
           'p_value': p_value}
    
```

```python
get_bootstrap(control, test)
```

по результатам бустрапа 0 не попадает в доверительный интервал, что говорит нам о том, что различия в группах статистически значимые
