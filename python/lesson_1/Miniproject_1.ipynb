1. Импортируйте библиотеку pandas как pd. Прочитайте датасет bookings.csv с разделителем ;. Ссылка на датасет находится в переменной path_to_file.


```python
import pandas as pd

bookings = pd.read_csv('2_bookings.csv', sep=';')

bookings
```

2. Как вы уже могли заметить, при сборе датасета произошла небольшая ошибка, и названия колонок были записаны в неудобном для дальнейшей работы формате. Исправьте ситуацию, приведя названия столбцов к нижнему регистру и заменив пробелы на знак нижнего подчеркивания. 
Например: Is Canceled -> is_canceled

```python
bookings = bookings.rename(str.lower, axis='columns')
bookings.columns = bookings.columns.str.replace(' ', '_')
bookings
```

3. Пользователи из каких стран совершили наибольшее число успешных бронирований? Бронирование считается успешным, если в дальнейшем не было отменено (переменная is_canceled). В качестве ответа выберите страны, входящие в топ-5.

```python
bookings_seccess = bookings.query("is_canceled == '0'")
bookings_seccess = bookings_seccess.country.value_counts()
bookings_seccess
```

4. На сколько ночей (stays_total_nights)  в среднем бронируют отели типа City Hotel? Resort Hotel? Запишите полученные значения в пропуски с точностью до 2 знаков после точки.

```python
stays_total_nights = bookings.groupby('hotel').agg({'stays_total_nights' : 'mean'}).sort_values('stays_total_nights').round(2)
stays_total_nights
```

5. Иногда тип номера, присвоенного клиенту (assigned_room_type), отличается от изначально забронированного (reserved_room_type). Такое может произойти, например, по причине овербукинга. Сколько подобных наблюдений встретилось в датасете?

*отмена бронирования также считается

```python
bookings_no_seccess = bookings.query("assigned_room_type != reserved_room_type")
bookings_no_seccess
bookings_no_seccess = bookings_no_seccess.reserved_room_type.value_counts().sum()
bookings_no_seccess
```

7. Посмотрите на числовые характеристики трёх колонок: adults, children и babies. Какая из них имеет наибольшее среднее значение?

```python
group_years = bookings.agg({'adults': 'mean'})
print(group_years)
group_years1 = bookings.agg({'children': 'mean'})
print(group_years1)
group_years2 = bookings.agg({'babies': 'mean'})
group_years2
```

8. Создайте колонку total_kids, объединив столбцы children и babies. Для отелей какого типа среднее значение переменной оказалось наибольшим?

```python
bookings['total_kids'] = bookings.children + bookings.babies
bookings_kids = bookings.groupby('hotel').agg({'total_kids': 'mean'}).round(2)
bookings_kids
```
